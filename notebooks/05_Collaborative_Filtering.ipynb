```python
from surprise import Dataset
from surprise import Reader
from surprise import SVD


import os
import csv
```

```python
def load_data(p_movies_file,p_ratings_file):
    reader = Reader(line_format= 'user item rating', sep=',',skip_lines=1, rating_scale=(0.5,5))
    ratings_dataset = Dataset.load_from_file(p_ratings_file, reader=reader)
    
    movieId_to_title = {}
    
    with open(p_movies_file, newline='',encoding= 'utf-8') as csvfile:
        movie_reader = csv.reader(csvfile)
        next(movie_reader)
        for row in movie_reader:
            movieID = int(row[0])
            movie_name = row[1]
            movieId_to_title[movieID]=movie_name
    return(ratings_dataset,movieId_to_title)

dataset , movieID_to_title = load_data("../processed/cleaned_movies.csv","../processed/cleaned_ratings.csv")

trainset = dataset.build_full_trainset()
```
